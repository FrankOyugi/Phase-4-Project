- The initial dataset comprises 272 columns representing different datetime columns, which can complicate data analysis and visualization tasks. To streamline this, we’ll transform the dataframe into a more manageable format where all datetime columns  are consolidated into a single column and their corresponding values in another. This process, known as “melting” the dataframe, enhances the ease of data manipulation and interpretation.

```python
def melt_data(df):
    # Melt the DataFrame 'df' into a long format
    melted = pd.melt(df, id_vars=['ZipCode', 'RegionID', 'SizeRank', 'City', 'State', 'CountyName', 'Metro', 'ROI'], var_name='Date')
    
    # Print unique values in the 'Date' column before conversion
    print("Unique values in 'Date' column before conversion:")
    print(melted['Date'].unique())
    
    # Identify non-date strings and filter them out
    valid_dates = pd.to_datetime(melted['Date'], errors='coerce', infer_datetime_format=True)
    invalid_dates = melted['Date'][valid_dates.isna()].unique()
    
    if len(invalid_dates) > 0:
        print("Non-date strings found in 'Date' column:")
        print(invalid_dates)
    
    # Drop rows with invalid date strings
    melted = melted[~melted['Date'].isin(invalid_dates)]
    
    # Convert the 'Date' column to datetime format
    melted['Date'] = pd.to_datetime(melted['Date'], infer_datetime_format=True)
    
    # Drop rows with missing values in the 'value' column
    melted = melted.dropna(subset=['value'])
    
    return melted

# Call the 'melt_data' function to process the DataFrame 'melted_df'
melted_df = melt_data(df_ca)


```

### Decomposition

We can decompose one of these zipcodes to evaluate the trend and seasonality.

```python
# Perform seasonal decomposition
decomposition = seasonal_decompose(ca_zips[california_zipcodes[3]], model='additive')

# Gather the trend, seasonality, and residuals
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot gathered statistics
plt.figure(figsize=(12, 12))

# Original data
plt.subplot(411)
plt.plot(ca_zips[california_zipcodes[3]], label='Original', color='blue')
plt.title('Original')
plt.legend(loc='best')

# Trend
plt.subplot(412)
plt.plot(trend, label='Trend', color='blue')
plt.title('Trend')
plt.legend(loc='best')

# Seasonality
plt.subplot(413)
plt.plot(seasonal, label='Seasonality', color='blue')
plt.title('Seasonality')
plt.legend(loc='best')

# Residuals
plt.subplot(414)
plt.plot(residual, label='Residuals', color='blue')
plt.title('Residuals')
plt.legend(loc='best')

plt.tight_layout()
plt.show()
```

We can see a clear upward trend with seasonality.

Now we can perform a quick stationarity check on the residual using the Dickey-Fuller test.

```python
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)

result = adfuller(ts_log_decompose)

print('P-Value:',result[1])
```

We can see from the p-value,that the time series is likely stationary,though we have seasonality.

## Stationarity Check

```python
def stationary_check(ts):
    roll_mean = ts.rolling(window=8, center=False).mean()
    roll_std = ts.rolling(window=8, center=False).std()
    
    fig = plt.figure(figsize=(12,8))
    plt.plot(ts,color='blue',label='Original')
    plt.plot(roll_mean,color='red',label='Rolling Mean')
    plt.plot(roll_std,color='black',label='Rolling Std')
    plt.legend()
    plt.show()
    
    result = adfuller(ts)

    print('Results of Dickey-Fuller test:')
    print('ADF Statistic:',result[0])
    print('P-Value:',result[1])
    print('Used Lag:',result[2])
    print('Number Of Observations Used:',result[3])
    print('Critical Values:',result[4])
```

```python
stationary_check(differenced_ts.dropna())
```

The time series is just about stationary.

## ARIMA

```python
def arima_forecast(ts, order_1, order_2, order_3):
    # Fit the ARIMA model
    model = ARIMA(ts, order=(order_1, order_2, order_3))
    fitted_model = model.fit()

    # Print the summary of the model
    print(fitted_model.summary())

    forecast_steps=len(test)
    # Generate forecasts
    forecast = fitted_model.forecast(steps=forecast_steps)
    
    # Calculate the Mean Squared Error (MSE)
    actual = ts[-forecast_steps:]  
    mse = mean_squared_error(actual, forecast[:len(actual)])
    
    # Plot the actual values and forecasts
    plt.figure(figsize=(12, 6))
    plt.plot(ts, label='Actual', color='blue')
    plt.plot(forecast, label='Forecast', color='red')
    plt.title('ARIMA Forecast')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend(loc='best')
    plt.show()
    
    # Print the MSE
    print(f'Mean Squared Error: {mse}')
    
    return fitted_model, forecast, mse
```

## Sarima Model

```python
def sarima_forecast(ts, order_1, order_2, order_3, seasonal_order_1, seasonal_order_2, seasonal_order_3, seasonal_order_4):
    # Fit the SARIMA model
    model = SARIMAX(ts, order= (order_1, order_2, order_3), seasonal_order= (seasonal_order_1, seasonal_order_2, seasonal_order_3, seasonal_order_4))
    fitted_model = model.fit()

    # Print the summary of the model
    print(fitted_model.summary())

    # Generate forecasts
    forecast_steps = len(test)
    forecast = fitted_model.forecast(steps=forecast_steps)
    
    # Calculate the Mean Squared Error (MSE)
    actual = ts[-forecast_steps:]  
    mse = mean_squared_error(actual, forecast[:len(actual)])
    
    # Plot the actual values and forecasts
    plt.figure(figsize=(12, 6))
    plt.plot(ts, label='Actual', color='blue')
    plt.plot(forecast, label='Forecast', color='red')
    plt.title('SARIMA Forecast')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend(loc='best')
    plt.show()
    
    # Print the MSE
    print(f'Mean Squared Error: {mse}')
    
    return fitted_model, forecast, mse
```

## Evaluation

#### After evaluating the Mean Squared Error values of both models,the best performing model is the Arima model as it has a lower MSE.

Now we can create a function which will iterate over the list of all the zipcodes in California,perfrom log transformation adn differencing,then fit them onto the best ARIMA model,then evaluate Alkaline Information Criterion values and use them to give us the top 5 and bottom 5 zipcodes to invest in.(This could take a while as we are iterating over 1288 columns which is computationally expensive).

```python
def arima_forecast(ts, order_1, order_2, order_3):
    # Fit the ARIMA model
    model = ARIMA(ts, order=(order_1, order_2, order_3))
    fitted_model = model.fit()
    
    # Suppress the printing of the model summary
    # print(fitted_model.summary())  # Commented out to avoid printing summaries

    forecast_steps = len(test)
    # Generate forecasts
    forecast = fitted_model.forecast(steps=forecast_steps)
    
    # Calculate the Mean Squared Error (MSE)
    actual = test  
    mse = mean_squared_error(actual, forecast[:len(actual)])
    
    return forecast, mse

def calculate_roi(forecast, actual, roi_dict):
    initial_value = actual.iloc[0]
    final_value = forecast.iloc[-1]
    roi = (final_value - initial_value) / initial_value * 100
    return roi

def evaluate_models(df, roi_dict, order_1=1, order_2=0, order_3=1, train_size=0.75):
    results = []

    for ts_name in df.columns:
        ts = df[ts_name].dropna()  # Drop NaNs from the time series
        split_idx = round(len(ts) * train_size)
        train = ts.iloc[:split_idx]
        global test  # To be used in the arima_forecast function
        test = ts.iloc[split_idx:]
        
        # Forecast with ARIMA
        forecast, mse = arima_forecast(train, order_1, order_2, order_3)
        
        # Get ROI from the dictionary
        actual_roi = roi_dict.get(ts_name, np.nan)
        
        if not np.isnan(actual_roi):
            # Calculate ROI based on forecast and actual ROI
            forecast_series = pd.Series(forecast, index=test.index)
            roi = calculate_roi(forecast_series, test, roi_dict)
            
            # Store results
            results.append((ts_name, roi, mse, actual_roi))
    
    # Convert results to DataFrame
    results_df = pd.DataFrame(results, columns=['TimeSeries', 'ForecastROI', 'MSE', 'ActualROI'])
    
    # Find top 5 models with highest Forecast ROI
    top_5_roi = results_df.nlargest(5, 'ForecastROI')
    
    # Find top 5 models with lowest Forecast ROI
    bottom_5_roi = results_df.nsmallest(5, 'ForecastROI')
    
    return top_5_roi, bottom_5_roi

# Create the ROI dictionary
roi_dict = melted_df.set_index('ZipCode')['ROI'].to_dict()


top_5_roi, bottom_5_roi = evaluate_models(ca_zips, roi_dict, order_1=1, order_2=0, order_3=1)

print("Top 5 Time Series with Highest Forecast ROI:")
print(top_5_roi)

print("\nBottom 5 Time Series with Lowest Forecast ROI:")
print(bottom_5_roi)
```
